```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('/Users/chenyalu/Desktop/IMT561/week5/homework5/ai_job_trends_dataset.csv')


top_10 = df.sort_values(by='Automation Risk (%)', ascending=False).head(10)
labels = top_10['Job Title'].values
values = top_10['Automation Risk (%)'].values
num_vars = len(labels)


angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()


fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
fig.patch.set_facecolor('#1e1e1e') # 深色背景，更有科技感
ax.set_facecolor('#1e1e1e')


ax.xaxis.grid(True, color='#444444', linestyle='--', alpha=0.7) # 经线
ax.yaxis.grid(True, color='#444444', linestyle='--', alpha=0.7) # 纬线

colors = plt.cm.Reds(values / 100) 
ax.scatter(angles, values, s=200, c=colors, edgecolors='white', zorder=10)

for angle, value in zip(angles, values):
    ax.plot([angle, angle], [0, value], color='red', alpha=0.3, linewidth=2)

ax.set_xticks(angles)
ax.set_xticklabels(labels, color='white', fontsize=11, fontweight='bold')

ax.set_ylim(0, 100)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels(['20%', '40%', '60%', '80%', '100%'], color='gray', fontsize=9)

plt.title("AI Automation Risk: Top 10 Jobs Distribution", color='white', size=18, pad=40)

plt.tight_layout()
plt.show()
```
